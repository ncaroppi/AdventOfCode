```python
import time

start_time = time.perf_counter()
```

```python
fp = '2025_day3_input.txt'

with open(fp, 'r') as file:
    input = file.readlines()

banks = []

for line in input:
    bank = line.strip('\n')
    banks.append(bank)
```

```python
def get_first_digit_and_index(bank):
    bank_size = len(bank)
    max_val = 0
    cell_index = 0

    for i in range(0, bank_size-1):
        if int(bank[i]) > max_val:
            max_val = int(bank[i])
            cell_index = i

    return max_val, cell_index

def get_second_digit(bank, i):
    bank_size = len(bank)
    second_val = 0
    for j in range(i+1, bank_size):
        if int(bank[j]) > second_val:
            second_val = int(bank[j])

    return second_val
    
def make_into_int(first_digit, second_digit):
    first_char = str(first_digit)
    second_char = str(second_digit)
    return(int(first_char + second_char))
```

```python
joltages = []

for bank in banks:
    first_digit, first_index = get_first_digit_and_index(bank)
    second_digit = get_second_digit(bank, first_index)
    joltage = make_into_int(first_digit, second_digit)
    joltages.append(joltage)

part1_time = time.perf_counter()

print(f'Part 1: {sum(joltages)}\nTime: {part1_time - start_time}')
```

```python
def add_next_digit(battery, next_digit):
    next_battery = ''
    for char in battery:
        next_battery += char

    next_battery += next_digit

    return next_battery   

def make_best_battery(battery):
    batt_len = len(battery)
    best_battery = battery[0:batt_len - 1]

    test_batt = ''
    for i in range(batt_len):
        test_batt = ''
        for j in range(batt_len):
            if i != j:
                test_batt += battery[j]

        if int(test_batt) > int(best_battery):
            best_battery = test_batt

 
    return best_battery
```

```python
BATT_SIZE = 12
test_bank = banks[0]
# test_bank = '818181911112111'

test_batt = test_bank[0:BATT_SIZE]
next_digit = test_bank[BATT_SIZE]

best_batteries = []

for bank in banks:
    test_batt = bank[0:BATT_SIZE]
    for i in range(BATT_SIZE,len(bank)):

        next_batt = add_next_digit(test_batt, bank[i])
        best_batt = make_best_battery(next_batt)
        test_batt = best_batt

    best_batteries.append(int(best_batt))

part2_time = time.perf_counter()

print(f'Part 2: {sum(best_batteries)}\nTime: {part2_time - part1_time}')
```
