```python
import time

start_time = time.perf_counter()
```

```python
fp = '2025_day3_input.txt'

with open(fp, 'r') as file:
    input = file.readlines()

banks = []

for line in input:
    bank = line.strip('\n')
    banks.append(bank)
```

```python
def get_first_digit_and_index(bank):
    bank_size = len(bank)
    max_val = 0
    cell_index = 0

    for i in range(0, bank_size-1):
        if int(bank[i]) > max_val:
            max_val = int(bank[i])
            cell_index = i

    return max_val, cell_index

def get_second_digit(bank, i):
    bank_size = len(bank)
    second_val = 0
    for j in range(i+1, bank_size):
        if int(bank[j]) > second_val:
            second_val = int(bank[j])

    return second_val

def get_nth_digit(bank, i, n):
    bank_size = len(bank)
    nth_val = 0
    cell_index = i
    for j in range(i+1, bank_size-n):
        if int(bank[j]) > nth_val:
            nth_val = int(bank[j])
            cell_index = j

    return nth_val, cell_index
    
def make_into_int(first_digit, second_digit):
    first_char = str(first_digit)
    second_char = str(second_digit)
    return(int(first_char + second_char))
```

```python
joltages = []

for bank in banks:
    first_digit, first_index = get_first_digit_and_index(bank)
    second_digit = get_second_digit(bank, first_index)
    joltage = make_into_int(first_digit, second_digit)
    joltages.append(joltage)

part1_time = time.perf_counter()

print(f'Part 1: {sum(joltages)}\nTime: {part1_time - start_time}')
```
