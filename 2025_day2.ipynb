```python
fp = '2025_day2_input.txt'

with open(fp, 'r') as file:
    input = file.readline()

sequence = input.split(',')

id_ranges = []

for vals in sequence:
    first_val, second_val = vals.split('-')
    id_ranges.append((int(first_val), int(second_val)))


```

```python
def even_count_of_digits(id_code):
    if len(str(id_code)) % 2 == 0:
        return True
```

```python
def is_invalid(id_code):
    if not even_count_of_digits(id_code):
        return False

    id_as_string = str(id_code)
    half_id_len = int(len(id_as_string)/2)

    if id_as_string[0:half_id_len] == id_as_string[half_id_len:]:
        return True

    return False
```

```python
invalid_ids = []

for id_range in id_ranges:
    for i in range(id_range[0],id_range[1]):
        if is_invalid(i):
            invalid_ids.append(i)

print(f'Part 1: {sum(invalid_ids)}')
```

```python
def is_all_substrings_identical(list_of_substrings):
    first_substring = list_of_substrings[0]
    for substring in list_of_substrings:
        if substring != first_substring:
            return False

    return True
```

```python
def get_substrings(full_id, substring_len):
    substrings = []
    id_len = len(full_id)
    for i in range(0, id_len, substring_len):
        substring = full_id[i: i+substring_len]
        substrings.append(substring)

    return substrings
```

```python
def get_substring_lengths(id_code):
    substring_lengths = []
    for i in range(1, len(id_code)):
        if len(id_code) % i == 0:
            substring_lengths.append(i)

    return substring_lengths
```

```python
def is_invalid_part_2(id_code):
    id_as_string = str(id_code)
    substring_lengths = get_substring_lengths(id_as_string)
    for length in substring_lengths:
        substrings = get_substrings(id_as_string, length)
        if is_all_substrings_identical(substrings):
            # print(f'{id_code} is invalid')
            return True

    # print(f'{id_code} is valid')
    return False


```

```python

invalid_ids = []

for id_range in id_ranges:
    for i in range(id_range[0],id_range[1]):
        if is_invalid_part_2(i):
            invalid_ids.append(i)

print(f'Part 2: {sum(invalid_ids)}')


```
