```python
import time
```

```python
start_time = time.perf_counter()

fp = '2025_day5_input.txt'

with open(fp, 'r') as file:
    input = file.readlines()

ranges = []
ingredients = []

for line in input:
    if '-' in line:
        range_low, range_high = line.split('-')
        ranges.append((int(range_low), int(range_high)))

    elif len(line) > 2:
        ingredients.append(int(line))
```

```python
def check_ingredient(ingredient_value, single_range):
    if ingredient_value >= single_range[0]:
        if ingredient_value <= single_range[1]:
            return True
        
    return False
```

```python
fresh_ingredients = 0

for ingredient in ingredients:
    for rng in ranges:
        if check_ingredient(ingredient, rng):
            fresh_ingredients += 1
            # print(f'ingredient: {ingredient} in {rng} \t fresh_ingredients = {fresh_ingredients}')
            break

time1 = time.perf_counter()

print(f'Part 1: {fresh_ingredients}, Time: {time1 - start_time}s')

```

```python
id_bounds = []
id_scores = {}

for rng in ranges:
    id_bounds.append(rng[0])
    id_bounds.append(rng[1])

id_bounds = sorted(list(set(id_bounds)))

for bound in id_bounds:
    id_scores[bound] = 0

for rng in ranges:
    id_scores[rng[0]] += 1
    id_scores[rng[1]] -= 1

running_score = 0
is_open = False

last_opener = id_bounds[0]

new_ranges = []

for bound in id_bounds:
    if not is_open:
        last_opener = bound

    running_score += id_scores[bound]
    if running_score > 0:
        is_open = True
    elif running_score <= 0:
        is_open = False
        new_ranges.append((last_opener, bound))

possible_IDs = 0

for rng in new_ranges:
    possible_IDs += rng[1] - rng[0] + 1

time2 = time.perf_counter()

print(f'Part 2: {possible_IDs} \tTime: {time2 - time1}')

```
