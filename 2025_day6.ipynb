```python
import time
import numpy as np
from math import prod
```

```python
start_time = time.perf_counter()

fp = '2025_day6_input.txt'

with open(fp, 'r') as file:
    lines = file.readlines()
    
    
rows = []

for line in lines:
	rows.append(line.strip().split())
      
```

```python
def perform_op(numbers) -> int:
	if numbers[-1] == '+':
		return (numbers[0] + numbers[1] + numbers[2] + numbers[3])
	else:
		return (numbers[0] * numbers[1] * numbers[2] * numbers[3])
```

```python
running_sum = 0

for i in range(len(rows[0])):
	numbers = (
		int(rows[0][i]),
		int(rows[1][i]),
		int(rows[2][i]),
		int(rows[3][i]),
		rows[4][i],
	)

	running_sum += perform_op(numbers)

time1 = time.perf_counter() - start_time

print(f'Part 1: {running_sum} Time = {time1:.4f}')
```

```python
def pad_numbers(numbers):

	max_len = 0

	for number in numbers:
		if len(number) > max_len:
			max_len = len(number)

	padded_numbers = []

	for number in numbers:
		
		zeros_to_add = max_len - len(number)
		new_number = number + ('0'*zeros_to_add)
		padded_numbers.append(new_number)

	print(f' max_len: {max_len} padded_numbers: {padded_numbers}')

	return (padded_numbers[0], padded_numbers[1], padded_numbers[2],padded_numbers[3], numbers[4])

```

```python
def find_result_add(set_of_numbers):
	result = sum(set_of_numbers)
	return result

def find_result_mult(set_of_numbers):
	result = prod(set_of_numbers)
	return result
```

```python
for i, line in enumerate(lines):
	lines[i] = line.strip()

rows = len(lines)
cols = len(lines[0])

digit_arr = np.zeros(shape=(rows, cols), dtype=str)


for i in range(cols):
	for j in range(rows):
		try:
			digit_arr[(j, i)] = lines[j][i]
		except:
			print(i, j)


digit_str = ''

digit_arr[(rows-1, cols-1)] = ' '

for i in range(cols-1, -1, -1):
	for j in range(rows):
		digit_str += digit_arr[(j, i)]
```

```python
op_set = []
number_str = ''

running_sum = 0

for i, char in enumerate(digit_str):
	try:
		if char == ' ':
			if number_str == '':
				pass
			else:
				op_set.append(int(number_str))
				# print(int(number_str))
			number_str = ''
		else:
			if char == '*':
				if number_str != '':
					op_set.append(int(number_str))
				number_str = ''
				result = prod(op_set)
				running_sum += result
				print(f'mult {op_set} \t result: {result} \t\t running: {running_sum}')
				op_set = []
			elif char == '+':
				if number_str != '':
					op_set.append(int(number_str))
				number_str = ''
				result = sum(op_set)
				running_sum += result
				print(f'add {op_set} \t result: {result} \t\t running: {running_sum}')
				op_set = []
			else:
				number_str += char
	except:
		print(f'ERROR! i: {i}  \t char: {char} \t num_string = {number_str}\t op_set{op_set}')

print(running_sum)
```
